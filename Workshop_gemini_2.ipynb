## Carga de datos

### Subtask:
Cargar el archivo `usuarios.csv` en un DataFrame de pandas.


**Reasoning**:
Import the pandas library and load the `usuarios.csv` file into a DataFrame named `df_usuarios`.



```python
import pandas as pd
df_usuarios = pd.read_csv('usuarios.csv')
```

## Exploración inicial

### Subtask:
Mostrar las primeras filas del DataFrame, los tipos de datos y un resumen estadístico básico para entender la estructura y el contenido de los datos.


**Reasoning**:
Display the first few rows, data types, and a statistical summary of the dataframe to understand its structure and content.



```python
display(df_usuarios.head())
df_usuarios.info()
display(df_usuarios.describe(include='all'))
```

## Análisis de variables categóricas

### Subtask:
Generar tablas de frecuencia y porcentajes para cada variable categórica (sexo, ciudad, barrio, red_preferida, contenido_preferido, ocupacion, pasatiempo).


**Reasoning**:
Generate frequency and percentage tables for each specified categorical column.



```python
categorical_cols = ['sexo', 'ciudad', 'barrio', 'red_preferida', 'contenido_preferido', 'ocupacion', 'pasatiempo']

for col in categorical_cols:
    print(f"\n--- Tabla de frecuencia para '{col}' ---")
    frequency_table = df_usuarios[col].value_counts()
    display(frequency_table)

    print(f"\n--- Tabla de porcentajes para '{col}' ---")
    percentage_table = (df_usuarios[col].value_counts() / len(df_usuarios)) * 100
    display(percentage_table)
```

## Análisis de variables categóricas

### Subtask:
Crear gráficos de barras horizontales para visualizar la distribución de cada variable categórica.


**Reasoning**:
Create horizontal bar charts for each categorical column to visualize their distributions.



```python
import plotly.express as px

categorical_cols = ['sexo', 'ciudad', 'barrio', 'red_preferida', 'contenido_preferido', 'ocupacion', 'pasatiempo']

for col in categorical_cols:
    value_counts = df_usuarios[col].value_counts()
    fig = px.bar(
        x=value_counts.values,
        y=value_counts.index,
        orientation='h',
        title=f'Distribución de {col}',
        labels={'x': 'Frecuencia', 'y': col}
    )
    fig.show()
```

## Análisis de variables categóricas

### Subtask:
Crear gráficos de pastel para visualizar la proporción de cada categoría en variables seleccionadas (por ejemplo, sexo, red_preferida).


**Reasoning**:
Import the plotly.express library and create pie charts for the selected categorical columns.



```python
import plotly.express as px

selected_categorical_cols = ['sexo', 'red_preferida']

for col in selected_categorical_cols:
    value_counts = df_usuarios[col].value_counts()
    fig = px.pie(
        values=value_counts.values,
        names=value_counts.index,
        title=f'Proporción de {col}'
    )
    fig.show()
```

## Análisis de variables numéricas

### Subtask:
Generar histogramas de frecuencia para las variables numéricas (edad, horas_de_uso).


**Reasoning**:
Generate histograms for the numerical columns 'edad' and 'horas_de_uso' using plotly express as instructed.



```python
numerical_cols = ['edad', 'horas_de_uso']

for col in numerical_cols:
    fig = px.histogram(df_usuarios, x=col, title=f'Distribución de {col}')
    fig.show()
```

## Análisis de variables numéricas

### Subtask:
Crear boxplots para visualizar la distribución y detectar posibles valores atípicos en las variables numéricas.


**Reasoning**:
Import the necessary library and define the numerical columns for creating boxplots to visualize distribution and identify outliers.



```python
import plotly.express as px

numerical_cols = ['edad', 'horas_de_uso']
```

**Reasoning**:
Iterate through the numerical columns and create a boxplot for each to visualize their distribution and identify outliers.



```python
for col in numerical_cols:
    fig = px.box(df_usuarios, y=col, title=f'Boxplot de {col}')
    fig.show()
```

## Análisis bivariado y multivariado

### Subtask:
Generar tablas de contingencia (por ejemplo, sexo vs red_preferida, ocupacion vs contenido_preferido).


**Reasoning**:
Create and display contingency tables for 'sexo' vs 'red_preferida' and 'ocupacion' vs 'contenido_preferido' using pd.crosstab.



```python
contingency_sexo_red = pd.crosstab(df_usuarios['sexo'], df_usuarios['red_preferida'])
print("Contingency Table: Sexo vs Red Preferida")
display(contingency_sexo_red)

contingency_ocupacion_contenido = pd.crosstab(df_usuarios['ocupacion'], df_usuarios['contenido_preferido'])
print("\nContingency Table: Ocupacion vs Contenido Preferido")
display(contingency_ocupacion_contenido)
```

## Análisis bivariado y multivariado

### Subtask:
Crear tablas dinámicas para explorar relaciones entre variables (por ejemplo, promedio de horas_de_uso por ocupación y red preferida).


**Reasoning**:
Create a pivot table to show the average of 'horas_de_uso' based on 'ocupacion' and 'red_preferida', then display the resulting pivot table.



```python
pivot_horas_ocupacion_red = df_usuarios.pivot_table(
    values='horas_de_uso',
    index='ocupacion',
    columns='red_preferida',
    aggfunc='mean'
)
display(pivot_horas_ocupacion_red)
```
